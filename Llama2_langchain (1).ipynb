# Import All the Required Libraries

```python
from langchain.llms import HuggingFacePipeline
```

```python
from transformers import AutoTokenizer
```

```python
import transformers
```

```python
import torch
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# Load the Llama 2 Model

```python
model="meta-llama/Llama-2-7b-chat-hf"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
```

```python
prompt = "What would be a good name for a company that makes colorful toys"
```

```python
print(llm(prompt))
```

```python
prompt = "I want to open a restaurant for Indian food. Suggest me a fancy name for this"
```

```python
print(llm(prompt))
```

# Prompt Templates

In many Large Language Model applications we donot pass the user input directly to the Large Language Model, we add the user input to a large piece of text called Prompt template.

```python
from langchain.prompts import PromptTemplate
```

```python
from langchain.chains import LLMChain
```

```python
prompt_template1 = PromptTemplate(input_variables = ["cuisine"],
                                  template = "I want to open a restaurant for {cuisine} food. Suggest a fancy name for this")
```

```python
input_prompt = prompt_template1.format(cuisine="Italian")
```

```python
print(input_prompt)
```

```python
chain = LLMChain(llm = llm, prompt = prompt_template1, verbose = True)
response = chain.run("Italian")
print(response)
```
